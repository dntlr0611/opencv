pip 명령을 이용하여 opencv-python 즉. Opencv 라이브러리를 다운받는다.



1.   먼저 pip install opencv-python을 명령어를 입력하여 opencv 라이브러리를 다운받습니다.

1.   pip install opencv-contrib-python을 명령어를 입력하여 opencv-contrib 라이브러리를 다운받습니다.

1.   pip install cmake 명령어를 입력하여서 dlib 라이브러리를 다운받기 전에 다운 받습니다.

2.   pip install dlib 명령어를 입력하여서 dlib 라이브러리를 다운받습니다.


  만약 오류가 발생한다면 https://pypi.python.org/simple/dlib 에 방문하셔서 최신의 내용의 whl 파일을 다운받고 명령어로 실행해주시기 바랍니다.
